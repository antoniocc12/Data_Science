 Ejemplo de particion de los datos (Train, test)

```python
#
X = [5, 80, 40, 24, 8] 
y = [105, 180, 140, 124, 108]

X_train = [40, 24, 80, 5] #para el conjunto de entrenamiento es importante que no sigan un orden especifico
y_train = [140, 124, 180, 105]

x = X_train[0]
x = 40

X_test = [8]
y_test = [108]
```

y representa la etiqueta o label de x. el test_size es el % que escogemos para que sea el test que suele ser el 20% pero eso lo escogemos nosotro

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

train_errors, va_errors = [], []
def plot_learning_curves(model, X, Y):
    # 80% de Train
    # 20% de Test
    X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.20)
    
    for m in range(1, len(X_train)):  # Incremental
        x = X_train[:m] #aqui el conjunto de entrenamiento vamos a ir cogiendolo poco a poco para entrenar el modelo
        # y es lo que vale o el valor que deberia dar cuando le das x la etiqueta 
        y = y_train[:m]
        model.fit(x, y) #la regresion lineal que le vamos a pasar, la vamos a entrenar por eso fit 
        # y train predict es el valor que ha dado o ha predicho al darle x 
        y_train_predict = model.predict(x) #aqui le estoy diciendo que me prediga
        # y val predict es el valor que ha dado al darle el conjunto no solo uno como la anterior
        y_val_predict = model.predict(X_val)#x_val es un valor que aun no ha sido entrenado

        #MEAN_SQUARED va a contener 
        train_errors.append(mean_squared_error(y, y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
        
        if m < 4:
            print("##### ITERACIÓN " + str(m) + " #####")
            print("x", x)
            print("y", y)
            print("y_train_predict", y_train_predict)
            print("-----\nX_val\n", X_val)
            print("y_val_predict", y_val_predict)

    plt.plot(train_errors, "r-+", linewidth=2, label="train")
    plt.plot(val_errors, "b-", linewidth=3, label="val")
```

```python
#  llamamos  auna funcion que va a recibir una funcion para x y para y 
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
plot_learning_curves(model=lin_reg, X=X, Y=y)
```

```python
# y = a + bX

# b, pendiente
b = lin_reg.coef_  # esto es sacar la pendiente 
b
```

```python
# a, secante
a = lin_reg.intercept_  # esto hacemos la secante, es decir donde corta . El linear regression es nuestro modelo que vamos a entrenar
a
```

```python
print("|||||||||||ERRORES||||||||||||")
print("train_errors\n\n", train_errors)
print("val_errors\n\n", val_errors)
```
