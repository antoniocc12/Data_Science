```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

train_errors, test_errors = [], []
def plot_learning_curves(model, X, Y):
    # 80% de Train
    # 20% de Test
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1111)
    
    for m in range(1, len(X_train)):  # Incremental
        x = X_train[:m]
        # y es el valor que debería dar cuando le das "x"
        y = y_train[:m]
        model.fit(x, y)

        # y_train_predict es el valor que ha predicho al darle "x"
        y_train_predict = model.predict(x)
        # y_val_predict es el valor que ha dado al darle "X_test"
        y_test_predict = model.predict(X_test)

        train_errors.append(mean_squared_error(y, y_train_predict))
        test_errors.append(mean_squared_error(y_test, y_test_predict))
        
        if m < 4:
            print("##### ITERACIÓN " + str(m) + " #####")
            print("x", x)
            print("y", y)
            print("model.coef_:", model.coef_)
            print("model.intercept_:", model.intercept_)
            print("y_train_predict", y_train_predict)
            print("-----\nX_test\n", X_test)
            print("y_test_predict", y_test_predict)

    plt.plot(train_errors, "r-+", linewidth=2, label="train")
    plt.plot(test_errors, "b-", linewidth=3, label="val")
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
plot_learning_curves(model=lin_reg, X=X, Y=y)
```

### Los pasos para resolver un ejercicio de regresión lineal

1. Cargamos el Conjunto de datos 
2. Partimos el conjunto en train y test
3. Creamos una variable que represente el modelo (inicializamos)
4. Fit (Entrenamos) el modelo con el conjunto Train
5. Calculamos el error de Train y de Test
6. ¿Es un buen error? --> Si no, repetir el proceso: 

    6.1 Cambiando la semilla 

    6.2 Eligiendo un % de test diferente

    6.3 Coger más datos
