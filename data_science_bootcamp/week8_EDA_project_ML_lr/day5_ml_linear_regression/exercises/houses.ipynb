### Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión lineal entre cada columna numérica con la columna 'price'. 

Cuando existe una columna(variable) que es la que se quiere comparar, se le llama 'target'. 

```python
#Una regresion simple es cada columna con price
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/USA_Housing.csv")
df
```

y = a + bX

```python
# Columns as Features
X = df[['Avg. Area Income']]
```

```python
# Price is my Target Variable, what we trying to predict
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
#importing the Linear Regression Algorithm
from sklearn.linear_model import LinearRegression
```

```python
#creating LinearRegression Object
lm = LinearRegression(n_jobs=-1)  # -1 to use all processors
```

```python
#Training the Data Model
lm.fit(X_train, y_train)
```

```python
#Displaying the Intercept
print(lm.intercept_)
```

```python
lm.coef_ 
```

```python
coeff_df = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
plt.scatter(X, y, color='b')
plt.plot(X, lm.predict(X), color='red')

plt.show()
```
