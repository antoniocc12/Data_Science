# 04 - Grouping

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users
```

```python
users.info()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

```python
users.groupby('occupation').mean()
```

```python
users.groupby('occupation')[['age']].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users['counter'] = np.ones(len(users))
```

```python
users.head()
```

```python
gender = users.groupby(['occupation', 'gender'])[['counter']].count().unstack()
gender
```

```python
gender['total'] = gender.apply(lambda x: sum(x), axis = 1)
```

```python
gender.head()
```

```python
gender['ratio_M'] = round(gender['counter']['M']/gender['total']*100, 2) 
gender['ratio_M']
```

```python
gender['ratio_M'] = gender['ratio_M'].apply(lambda x: str(x) + '%')
```

```python
gender.loc["doctor", "ratio_M"] = np.nan
```

```python
gender[['ratio_M']]
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
# aggregate es lo mismo que poner agg
users.groupby('occupation')['age'].agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])[['age']].mean()
```
