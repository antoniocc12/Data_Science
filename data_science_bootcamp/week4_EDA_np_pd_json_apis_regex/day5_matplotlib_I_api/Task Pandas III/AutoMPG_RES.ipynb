# 06 - Combining datasets

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the datasets you'll find in the folder `data` and assign each to a variable called cars1 and cars2

The following exercise uses data from [UC Irvine Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Auto+MPG)

```python
cars1 = pd.read_csv("data/cars1.csv")
cars2 = pd.read_csv("data/cars2.csv")

display(cars1.head())
cars2.head()
```

### Step 3. Have a look at the columns of the dataset.

```python
print(cars1.columns)
print(cars2.columns)
```

### Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
cars1.isnull().all()
```

```python
#1
deleted_columns = cars1.drop(columns=['Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13'], axis=1)
deleted_columns
```

```python
#2
cars1.drop(cars1.columns[cars1.columns.str.contains('Unnamed', case=False)], axis=1, inplace=True)
cars1
```

```python
#3
cars1 = cars1.loc[:, "mpg":"car"]
cars1.head()
```

```python
#4
cars_copy = cars1.dropna(axis="columns", how="all")
cars_copy
```

### Step 5. What is the number of observations in each dataset?

```python
print(cars1.shape[0])
print(cars2.shape[0])
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
concated = pd.concat([cars1, cars2], axis = 0)
concated
```

```python
merged = pd.merge(cars1, cars2, how="outer")
merged
```

```python
cars = cars1.append(cars2)
cars
```

### Step 7. Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
owners = pd.Series(np.random.randint(15000, 73001, size=len(cars)), name = "owners")
owners
```

### Step 8. Add the column owners to cars

```python
cars.reset_index(drop = True, inplace = True)
```

```python
pd.concat([cars, owners], axis =1)
```
